Was ist der durchschnittliche Preis mit Bezug auf den „body-style“ eines Autos?

```python
df = pd.read_csv('data/automobile.csv')
```

```python
df.head()
```

```python
style_price = df[['body-style', 'price']].groupby('body-style').mean()
style_price
```

Welches sind die drei „body-styles“ mit den durchschnittlich höchsten Preisen?

```python
style_price.sort_values('price', ascending=False).head(3)
```
